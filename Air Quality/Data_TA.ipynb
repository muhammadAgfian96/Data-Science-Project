```python
def show_plot(col, tanggal,r_max, r_min, step=10):
    kondisi = df['date']<'2019-10-'+str(tanggal+1)
    konidisi_2 =df['date']>'2019-10-'+str(tanggal)
    
    grafik = sns.lineplot(x=df[kondisi & konidisi_2]['date'], 
                 y=df[kondisi & konidisi_2][col],
                 palette='asd'
                )
    
    if (df[kondisi & konidisi_2][col].max() > r_max or df[kondisi & konidisi_2][col].min() < r_min):
        grafik
    else:
        grafik.set_yticks(np.arange(r_min, r_max, step))
    plt.xticks(rotation=90)
    plt.title("{col} {tanggal}"
              .format(col=col, 
                      tanggal='2019-10-'+str(tanggal)), 
            fontdict={
              'fontweight':'bold',
            'fontsize': 15},
              pad=5
             )
```

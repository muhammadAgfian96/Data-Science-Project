**think how to fill nan value bathrooms and beds**

okay, i will use median based on **property_type** and __room_type__ to fill na value in **'beds'** and **'bathrooms'**. Why? because from that both parameter, usually we can know how much beds and bathrooms inside.

```python
for_fill_df = df.groupby(['property_type','room_type']).median()[['beds','bathrooms']]
for_fill_df
```

```python
idx_high = for_fill_df.index.levels[0]
idx_low = for_fill_df.index.levels[1]

property_tp = 'Aparthotel'
room_tp = 'Private room'
#0 -> beds
#1 -> bathroom
for_fill_df.loc[property_tp, room_tp][0]
```

```python
def fill_bathrooms(x):
    col_fill = 1 # because bathrooms
    if (pd.isnull(x['bathrooms']) == True):
        property_tp = x['property_type']
        room_tp = x['room_type']
        isi = for_fill_df.loc[property_tp, room_tp][col_fill]
        return isi
    else:
        return x['bathrooms']
```

```python
df['bathrooms'] = df[['property_type', 'room_type', 'bathrooms']].apply(fill_bathrooms, axis=1)
```

```python
# check bathrooms null value
df['bathrooms'].isnull().sum()
```

```python
def fill_beds(x):
    col_fill = 0 # because it is beds
    if (pd.isnull(x['beds']) == True):
        property_tp = x['property_type']
        room_tp = x['room_type']
        isi = for_fill_df.loc[property_tp, room_tp][col_fill]
        return isi

    else:
        return x['beds']
```

```python
df['beds'] = df[['property_type', 'room_type', 'beds']].apply(fill_beds, axis=1)
```

```python
# check bathrooms null value
df['beds'].isnull().sum()
```
